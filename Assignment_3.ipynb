Q.1 What is the difference between a function and a method in Python?
- Function (Theoretical Definition)
A function in Python is a self-contained block of reusable code that performs a specific task. It is defined using the def keyword and can be invoked independently. Functions can accept input arguments and return output values. They are part of Python's procedural programming paradigm and do not inherently operate on any object unless explicitly passed.

 Method (Theoretical Definition)
A method is a function that is associated with an object. In Python, methods are defined within a class and are meant to operate on instances of that class (objects). They implicitly take the instance (self) as their first argument and are part of Python's object-oriented programming paradigm. Methods are invoked on objects and can access and modify the objectâ€™s internal state.

Example of Function

```python
def add(a, b):
    return a + b

# Calling the function
result = add(3, 5)
print(result)  # Output: 8
```

Example of a Method

```python
class Calculator:
    def add(self, a, b):
        return a + b

# Creating an object of the class
calc = Calculator()

# Calling the method
result = calc.add(3, 5)
print(result)  # Output: 8
```

Q.6 Explain the concept of generators in Python and how they are defined.
- ðŸ”¹ Concept of Generators:

 A generator is a special type of function in Python that returns an iterator and allows you to iterate over a sequence of values lazily (one value at a time, only when needed).
Instead of using return, it uses the yield keyword to produce a value and pause the functionâ€™s execution, resuming from that point on the next call.

- ðŸ”¹ Why Use Generators?

 Memory-efficient: Does not store the entire sequence in memory.

 - Useful for large or infinite data streams.

 - Faster startup time than building entire lists.

- ðŸ”¹ How to Define a Generator:

 Generators are defined like normal functions but use yield instead of return.

 Example ;

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

gen = count_up_to(3)

for num in gen:
    print(num)
```

Q.7  What are the advantages of using generators over regular functions?
- Generators provide several key advantages compared to regular functions that return complete data structures:

 1.Memory Efficiency:

 Generators produce items one at a time and only when requested, so they do not store the entire sequence in memory. This makes them ideal for handling large datasets or infinite sequences.

 2.Lazy Evaluation:

 Values are generated on demand rather than computed all at once. This improves performance and reduces initial loading time.

 Ability to Represent Infinite Sequences:
Because generators yield items one by one, they can generate infinite data streams without running out of memory.

 Improved Performance:
By avoiding the creation of full lists, generators can be faster and more efficient in iterations over large data.

 Simplified Code:
Using yield allows writing cleaner, more readable code for producing sequences compared to managing and returning lists manually.

Q.9  Explain the purpose and usage of the `map()` function in Python.
- ### Purpose and Usage of the `map()` Function in Python (Theoretical)

 The `map()` function in Python is used to apply a specified function to **each item** of one or more iterables (such as lists or tuples) and returns an iterator that yields the results. It enables functional-style programming by allowing transformation or processing of data sequences without explicit loops.

 `map()` takes a function and one or more iterables as arguments. The function is applied to the items of the iterable(s) in parallel, and the results are produced lazily as a map object. This map object can be converted to other iterable types like lists or tuples if needed.

 This function is useful for concise, readable, and efficient data processing tasks, especially when combined with anonymous functions (lambdas).

 Example ;


```python
numbers = [1, 2, 3, 4]
squared = map(lambda x: x * x, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]
```

Q.10 What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- ### Difference Between `map()`, `filter()`, and `reduce()` Functions (Theory)

* **`map()`** applies a given function to every item of one or more iterables and returns an iterator with the transformed results. It is used to **transform** data.

* **`filter()`** applies a function that returns a boolean value (True or False) to each item of an iterable and returns an iterator containing only those items for which the function returns True. It is used to **select** data based on a condition.

* **`reduce()`** (from the `functools` module) applies a binary function cumulatively to the items of an iterable, reducing the iterable to a single accumulated value. It is used to **aggregate** data into one result.

In summary, `map()` transforms each item, `filter()` selects items, and `reduce()` combines items to produce a single output.

Q.1 . Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

# Example usage:
nums = [1, 2, 3, 4, 5, 6]
print(sum_even_numbers(nums))
```

Q.4  Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2 or n > 200:
        return False  # Outside the valid range
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage:
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number.")
```

Q.5  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.count += 1
            self.a, self.b = self.b, self.a + self.b
            return self.a

# Example usage:
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

Q.6  Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp

# Example usage:
for power in powers_of_two(5):
    print(power)
```

Q.7  Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
for line in read_file_line_by_line('example.txt'):
    print(line)
```

Q.9 Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Conversion formula: F = C * 9/5 + 32
fahrenheit_temps = list(map(lambda c: c * 9/5 + 32, celsius_temps))

print("Celsius:", celsius_temps)
print("Fahrenheit:", fahrenheit_temps)
```

Q.10 Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, input_str))

# Example usage
text = "Hello, World!"
result = remove_vowels(text)

print("Original:", text)
print("Without vowels:", result)
```
